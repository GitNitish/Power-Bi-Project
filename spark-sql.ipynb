```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark=SparkSession.builder.appName("Basics").getOrCreate() 
```

Below code is to read a specific file and i used header option to true to make a proper table with columns name out of it.

```python
df=spark.read.csv("/kaggle/input/iris/Iris.csv",header=True)  
df.show(5)

df.createOrReplaceTempView("flowers")

count= spark.sql("Select count(*) as row_count from flowers")
count.show()
```

```python
view=spark.sql("select distinct Species from flowers")
view.show()
```

```python
sepal=spark.sql("select MAX(SepalLengthCm)as max_sepal_length,MIN(SepalLengthCm)as min_sepal_length, MAX(SepalWidthCm)as max_sepal_width, MIN(SepalWidthCm)as min_sepal_width, Species from flowers group by Species")
sepal.show()

petal=spark.sql("select MAX(PetalLengthCm)as max_petal_length,MIN(PetalLengthCm)as min_petal_length, MAX(PetalWidthCm)as max_petal_width, MIN(PetalWidthCm)as min_petal_width, Species from flowers group by Species")
petal.show()
```
